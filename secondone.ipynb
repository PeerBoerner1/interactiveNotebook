## Ask the user for input

```pyspark
%local import ipywidgets as z
from IPython.display import display
import pandas as pd
```

```pyspark
%%local filePath = z.Dropdown(
    options=['s3://amazon-reviews-pds/parquet/product_category=Books/*.parquet', 'file://something/local/file.txt', 'c://some/dir/anotherfile.txt'],
    description='Select a file:',
    disabled=False,
)
minStarRating = z.BoundedIntText(
    min=0,
    max=5,
    step=1,
    description='Min Rating:',
    disabled=False
)
maxStarRating = z.BoundedIntText(
    min=0,
    max=5,
    value=5,
    step=1,
    description='Max Rating:',
    disabled=False
)
fromDate = z.DatePicker(
    description='From Date',
    disabled=False
)
display(filePath, minStarRating, maxStarRating, fromDate)
```

```pyspark
%%local print(f"File is: {filePath.value}\nMin star rating is: {minStarRating.value}\nMax Star rating is: {maxStarRating.value}\nFrom date is: {fromDate.value}")
```

```pyspark
%%local 
fname = filePath.value
minStar = str(minStarRating.value)
maxStar = str(maxStarRating.value)
frDate = str(fromDate.value)
```

```pyspark
%%send_to_spark -i fname
```

```pyspark
%%send_to_spark -i minStar
```

```pyspark
%%send_to_spark -i maxStar
```

```pyspark
%%send_to_spark -i frDate
```

```pyspark
%%send_to_spark -i fname -t str
```

## Load some big data

```pyspark
# Load the amazon book sample data source ~20 million rows
spdf = spark.read.parquet(fname)
```

```pyspark
spdf.printSchema()
```

```pyspark
# Grab a sub-set of all the fields
# Other analytic transformation would happen here as well.
votesDf=spdf.select('customer_id', 'review_date','product_title','star_rating', 'total_votes').sort('customer_id')
print(f"Number of rows in spark data frame: {votesDf.count():,}")
```

```pyspark
%%spark -o votesDf
```
