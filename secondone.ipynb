## Load some big data

```pyspark
# Load the amazon book sample data source ~20 million rows
spdf = spark.read.parquet('s3://amazon-reviews-pds/parquet/product_category=Books/*.parquet')
```

```pyspark
spdf.printSchema()
```

```pyspark
# Grab a sub-set of all the fields
# Other analytic transformation would happen here as well.
votesDf=spdf.select('customer_id', 'review_date','product_title','star_rating', 'total_votes').sort('customer_id')
print(f"Number of rows in spark data frame: {votesDf.count():,}")
```

```pyspark
%%spark -o votesDf
```

```pyspark
%%local votesDf['review_date'].where(votesDf.review_date > '2014-07-12').count()
```
